```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from cf_matrix import make_confusion_matrix
sns.set_context('talk')
```

```python
cf_matrix = np.array([[55,8],
                      [6,38]])
labels = ['True Neg','False Pos','False Neg','True Pos']
categories = ['Healthy', 'Heart-failure']
make_confusion_matrix(cf_matrix, group_names=labels, categories=categories,figsize=(5,5),cmap="YlGnBu",
                       title='Mitral fold-1 with ResNet-18',sum_stats=False,cbar=False,xyplotlabels=False)
plt.savefig('ResNet.png', dpi=600)
```
